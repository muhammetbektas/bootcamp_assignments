```
import numpy as np
import matplotlib.pyplot as plt
import time
from sklearn.datasets import fetch_openml
import umap
from sklearn.preprocessing import StandardScaler
```

```
mnist = fetch_openml('Fashion-MNIST', version=1, cache=True)
```

```
np.random.seed(123)

indices=np.random.choice(70000,10000)
X=mnist.data[indices] / 255.0
y=mnist.target[indices]
y=[int(i) for i in y]
```

```
scaler=StandardScaler()
X_scl=scaler.fit_transform(X)
```

```
time_start = time.time()

umap_results = umap.UMAP(n_neighbors=5,
                      min_dist=0.3,
                      metric='correlation').fit_transform(X)
print('UMAP is complicated! elapsed time: {} second'.format(time.time()-time_start))
```

```
plt.scatter(umap_results[:,0],umap_results[:,1],c=y)
plt.show()
```

Umap is the best in clustering method than t-SNE and PCA for this problem. But not enough to distinguish certainly.


```
umap_results = umap.UMAP(n_neighbors=8,
                      min_dist=0.3,
                      metric='correlation').fit_transform(X)
plt.scatter(umap_results[:,0],umap_results[:,1],c=y)
plt.show()
```

```
umap_results = umap.UMAP(n_neighbors=5,
                      min_dist=0.5,
                      metric='correlation').fit_transform(X)
plt.scatter(umap_results[:,0],umap_results[:,1],c=y)
plt.show()
```

We have tried three combination with UMAP. We have changed the hyperparameter values. We can say, second combination is the best.
