```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_openml
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
mnist = fetch_openml('Fashion-MNIST', version=1, cache=True)
```

```python
np.random.seed(123)

indices = np.random.choice(70000,10000)
X = mnist.data[indices] / 255.0
y = mnist.target[indices]
y=[int(i) for i in y]
```

```python
scaler = StandardScaler()
X_std = scaler.fit_transform(X)
```

```python
pca = PCA(n_components=2)
pca_components = pca.fit_transform(X_std)
```

```python
plt.figure(figsize=(10,5))
plt.scatter(pca_components.T[0],pca_components.T[1],c=y)
plt.show()
```

As we can see above, it is impossible to distinguish easily. Clusters are complicatedly. 
We conclude that 2-components data is not sufficient to observe classes.
