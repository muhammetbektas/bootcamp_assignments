```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.animation as animation
%matplotlib notebook
%matplotlib inline
```

```python
orneklem1=[]
orneklem2=[]
for i in range(50):
    a=np.random.normal(2.5,45,1000)
    orneklem1.append(np.mean(a))
for j in range(1000):
    b=np.random.normal(4,10,1000)
    orneklem2.append(np.mean(b))
plt.figure(figsize=(8,5))

plt.hist(orneklem1,alpha=0.5,label='Örneklem 1')
plt.hist(orneklem2,alpha=0.5,label='Örneklem 2')
plt.legend(loc='upper right')
plt.show()
```

```python
from scipy import stats
plt.figure(figsize=(10,4),dpi=100)
plt.subplot(1,2,1)
baslik_font = {'family': 'arial', 'color': 'darkred','weight': 'bold','size': 13 }
res=stats.probplot(orneklem1,plot=plt)
plt.title('Q-Q Çizgesi',fontdict=baslik_font)
plt.xlabel("Teorik Değerler", fontdict = baslik_font)
plt.ylabel("Örnek Veriler", fontdict = baslik_font)

plt.subplot(1,2,2)
res=stats.probplot(orneklem2,plot=plt)
plt.title('Q-Q Çizgesi',fontdict=baslik_font)
plt.xlabel("Teorik Değerler", fontdict = baslik_font)
plt.ylabel("Örnek Veriler", fontdict = baslik_font)
plt.show()
```

```python
ortalama=[]

for i in range(1, 10000, 10):
    atis=np.random.binomial(1, 0.75, i)
    ortalama=np.append(ortalama, np.mean(atis))
    
plt.plot(ortalama[1:])
plt.axhline(0.75, color="r")
plt.show()
```
