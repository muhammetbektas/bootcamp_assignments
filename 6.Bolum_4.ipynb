```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from sklearn import linear_model
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

### Soru 2

```python
sicaklik=pd.read_csv('../weatherHistory.csv')
sicaklik.head()
```

```python
y=sicaklik['Apparent Temperature (C)']-sicaklik['Temperature (C)']
x=sicaklik[['Humidity','Wind Speed (km/h)']]
```

```python
lrm=linear_model.LinearRegression()
lrm.fit(x,y)
```

```python
print('Degiskenler: \n', lrm.coef_)
print('Sabit Deger (bias): \n',lrm.intercept_)
```

```python
x=sm.add_constant(x)
results=sm.OLS(y,x).fit()
results.summary()
```

#### y=2.4381 - 3.0292 x Humidity - 0.1193 x Wind Speed (km/h)

Katsayılar istatistiksel olarak anlamlıdır.Çünkü; p değerleri 0.05 den küçüktür. Hissedilen sıcaklık nem arttıkça artması gerekirken Humidity ile hissedilen sıcaklık ve gerçek sıcaklık farkı arasında negatif bir ilişki vardır. Bunu şurdan anlıyoruz çünkü; Humidity' nin katsayısı -3.0292 dir. Rüzgar hızının etkisi hava sıcaklığı ve hız değerine göre değiştiği için anlamlı bir yorum yapamıyorum.Ancak tahmini katsayılar önceki beklentilerim doğrultusunda değil.

R-squared değeri Humidity ve Wind Speed (km/h) bağımsız değişkenleri ile hedef(bağımlı) değişkeni arasındaki korelasyon değerinin karesini göstermektedir.0.288 değerindeki R-squared bize hedef değişkeni içindeki varyasyonun (çeşitliliğin) %28.8’inin  Humidity ve Wind Speed (km/h) değişkenlerine atfedilebileceğini söylemektedir.

Nem miktarındaki 1 birimlik artış hissedilen sıcaklık ile gerçek sıcaklık farkını 3.0292 (C) azaltacaktır. Bu ilişki anlamlıdır(p_value<0.05).p değeri sıfıra eşittir.
Rüzgar hızındaki 1 birimlik artış ise hissedilen sıcaklık ile gerçek sıcaklık farkını 0.1193 (C) azaltacaktır. Bu ilişki anlamlıdır(p_value<0.05).p değeri sıfıra eşittir.
Hedef ve açıklayıcı değişkenler arasında negatif ilişki vardır.

```python
y=sicaklik['Apparent Temperature (C)']-sicaklik['Temperature (C)']
sicaklik['hum_x_wind']=sicaklik['Humidity']*sicaklik['Wind Speed (km/h)']
x=sicaklik[['Humidity','Wind Speed (km/h)','hum_x_wind']]
```

```python
lrm=linear_model.LinearRegression()
lrm.fit(x,y)
```

```python
print('Degiskenler: \n', lrm.coef_)
print('Sabit Deger (bias): \n',lrm.intercept_)
```

```python
x=sm.add_constant(x)
results=sm.OLS(y,x).fit()
results.summary()
```

#### y=0.0839+0.1775 x Humidity + 0.0905 x Wind Speed (km/h) - 0.2971 x hum_x_wind
Katsayılar istatistiksel olarak anlamlı. Katsayılar değişti ve hedef değişken ile açıklayıcı değişkenler(Humidity,Wind Speed (km/h)) arasındaki ilişki pozitif oldu.

### Soru 3

```python
house=pd.read_csv('../trainn.csv')
house.head()
```

Ön veri araştırması sonucu hedef değişkenimizi tahmin etmek için kullanacağım bir kaç özellik belirledim.


```python
housee=house[['MSSubClass','Neighborhood','Condition2',
                          'OverallCond','Exterior1st','ExterCond','Heating','SaleType','SalePrice']]
housee.head()
```

```python
SubClassDict = {20: "1-STORY 1946 & NEWER ALL STYLES", 
        30: "1-STORY 1945 & OLDER", 
        40: "1-STORY W/FINISHED ATTIC ALL AGES", 
        45: "1-1/2 STORY - UNFINISHED ALL AGES", 
        50: "1-1/2 STORY FINISHED ALL AGES", 
        60: "2-STORY 1946 & NEWER", 
        70: "2-STORY 1945 & OLDER", 
        75: "2-1/2 STORY ALL AGES", 
        80: "SPLIT OR MULTI-LEVEL", 
        85: "SPLIT FOYER", 
        90: "DUPLEX - ALL STYLES AND AGES", 
       120: "1-STORY PUD (Planned Unit Development) - 1946 & NEWER", 
       150: "1-1/2 STORY PUD - ALL AGES", 
       160: "2-STORY PUD - 1946 & NEWER", 
       180: "PUD - MULTILEVEL - INCL SPLIT LEV/FOYER", 
       190: "2 FAMILY CONVERSION - ALL STYLES AND AGES"}
```

```python
OverallCondDict={10:"Very Excellent",
       9:"Excellent",
       8:"Very Good",
       7:"Good",
       6:"Above Average",
       5:"Average",
       4:"Below Average",
       3:"Fair",
       2:"Poor",
       1:"Very Poor"}
```

```python
housee.MSSubClass = housee.MSSubClass.astype('category')
housee.OverallCond= housee.OverallCond.astype('category')
```

```python
housee.MSSubClass.map(SubClassDict).head()
```

```python
housee.OverallCond.map(OverallCondDict).head()
```

```python
endocing=pd.concat([housee,pd.get_dummies(housee[['Neighborhood','Condition2',
                                        'Exterior1st','ExterCond','Heating','SaleType','MSSubClass','OverallCond']])],axis=1).drop(columns=['Neighborhood','Condition2',
                                        'Exterior1st','ExterCond','Heating','SaleType','MSSubClass','OverallCond'],axis=1)
```

```python
y=endocing['SalePrice']
x=pd.get_dummies(housee[['Neighborhood','Condition2',
                                        'Exterior1st','ExterCond','Heating','SaleType','MSSubClass','OverallCond']])
```

```python
import statsmodels.api as sm
pd.options.display.float_format='{:,.7f}'.format
x = sm.add_constant(x)
results = sm.OLS(y, x).fit()
results.summary()
```

```python
results.pvalues[results.pvalues<0.05].index.values
```

```python
anlamlilar=pd.concat([endocing.SalePrice,endocing[results.pvalues[results.pvalues<0.05].index.drop('const')]],axis=1)
```

```python
y=anlamlilar['SalePrice']
x=anlamlilar.drop(columns='SalePrice')
import statsmodels.api as sm
x=sm.add_constant(x)
result=sm.OLS(y,x).fit()
result.summary()
```

Anlamli olan özellikler ile çalıştığımda R-squared'im 0.64 oldu.İlk duruma göre azaldı.İyileştirme olmadı.

Aşağıda anlamli olan özelliklerin bir birim değiştiğinde SalePrice'ı kaç birim etkileyeceğini belirlediğimiz bu özelliklere ait katsayıları görüyoruz.

```python
result.params[result.pvalues<0.05]
```

Yukaridaki ozellikler SalePrice üzerinde belirgin etkiye sahiptir.
Anlamlilar ile çalıştığımda modele giren kısıtlar 87 den 37 ye düştü. Modelim iyileşti. R-squared im %64 e düştü. Düşmesinin nedeni ise p_value 0.05 e yakın aslında anlamlı denebilecek(0.06 gibi sıfıra yakın) değerler vardı. Bu değerleride attığımızda R-squared azaldı ancak çok düşük miktarda. Sonuç mantıklı.
