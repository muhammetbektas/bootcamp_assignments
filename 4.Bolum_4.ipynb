```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
edu=pd.read_csv('../states_all.csv')
```

```python
edu3=edu.copy()
columns =   ["ENROLL", "TOTAL_REVENUE", "FEDERAL_REVENUE",
             "STATE_REVENUE", "LOCAL_REVENUE", "TOTAL_EXPENDITURE",
             "INSTRUCTION_EXPENDITURE", "SUPPORT_SERVICES_EXPENDITURE",
             "OTHER_EXPENDITURE", "CAPITAL_OUTLAY_EXPENDITURE", "GRADES_PK_G",
             "GRADES_KG_G", "GRADES_4_G", "GRADES_8_G", "GRADES_12_G", "GRADES_1_8_G",
             "GRADES_9_12_G", "GRADES_ALL_G"]

for col in columns:
     edu3[col] = edu3.groupby('STATE')[col].transform(lambda x: x.interpolate(limit_direction = 'both'))
```

```python
edu3[edu3.STATE == "ALABAMA"]
```

```python
edu3
```

```python
plt.figure(figsize=(15,5))

plt.subplot(1,2,1)
plt.hist(edu3["TOTAL_REVENUE"])
plt.title("total revenue")

plt.subplot(1,2,2)
plt.hist(edu3['TOTAL_EXPENDITURE'])
plt.title("total expenditure")
plt.show()
```

```python
plt.rcParams['figure.dpi'] = 100

plt.figure(figsize=(15,5))

plt.subplot(1,2,1)
plt.boxplot(edu3['TOTAL_REVENUE'].dropna())
plt.title("total revenue")

plt.subplot(1,2,2)
plt.boxplot(edu3['TOTAL_EXPENDITURE'].dropna())
plt.title("total expenditure")
plt.show()
```

```python
plt.figure(figsize=(15,5))

plt.subplot(1,2,1)
plt.boxplot(edu3['TOTAL_REVENUE'].dropna(),whis=7)
plt.title("total revenue")

plt.subplot(1,2,2)
plt.boxplot(edu3['TOTAL_EXPENDITURE'].dropna(),whis=5)
plt.title("total expenditure")
plt.show()
```

```python
from scipy.stats.mstats import winsorize

winsorize_revenue=winsorize(edu3['TOTAL_REVENUE'].dropna(),(0,0.05))
winsorize_expenditure=winsorize(edu3['TOTAL_EXPENDITURE'].dropna(),(0,0.05))

winsorize_revenue
winsorize_expenditure
```

```python
plt.figure(figsize = (15,5))
plt.subplot(1,2,1)
plt.boxplot(winsorize_revenue,whis=5)
plt.title('Total Revenue kutu grafigi')

plt.subplot(1,2,2)
plt.boxplot(winsorize_expenditure,whis=5)
plt.title('Total Expenditure kutu grafigi')
plt.show()
```

```python
fark=edu3.TOTAL_REVENUE-edu3.TOTAL_EXPENDITURE
plt.figure(figsize = (5,5))
plt.boxplot(fark.dropna(),whis=5)
plt.show()
```

```python
winsorize_fark=winsorize(fark.dropna(),(0.05,0.05))
plt.boxplot(winsorize_fark,whis=5)
plt.show()
```

```python
fark2=winsorize_revenue-winsorize_expenditure
```

```python
plt.boxplot(fark2,whis=5)
plt.show()
```

```python
winsorize_fark2=winsorize(fark2,(0.05,0.05))
plt.boxplot(winsorize_fark2,whis=5)
plt.title('Butce farki2 kutu grafigi')
```

```python
pd.DataFrame({'Winsorize butce':winsorize_fark,'Winsorize butce2':winsorize_fark2}).describe()
```
