```python
import numpy as np
import matplotlib.pyplot as plt
import time
from sklearn.datasets import fetch_openml
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler
```

```python
mnist = fetch_openml('Fashion-MNIST', version=1, cache=True)
```

```python
np.random.seed(123)

indices=np.random.choice(70000,10000)
X=mnist.data[indices] / 255.0
y=mnist.target[indices]
y=[int(i) for i in y]
```

```python
scaler=StandardScaler()
X_scl=scaler.fit_transform(X)
```

```python
time_start=time.time()
tsne=TSNE(n_components=2,verbose=1,perplexity=5,n_iter=300)
tsne_results=tsne.fit_transform(X_scl)
print('t-SNE is completed, elapsed time is {}'.format(time.time()-time_start))
```

```python
plt.figure(figsize=(10,5))
plt.scatter(tsne_results[:, 0], tsne_results[:, 1],c=y)
plt.xticks([])
plt.yticks([])
plt.axis('off')
plt.show()
```

```python
time_start=time.time()
tsne=TSNE(n_components=2,verbose=1,perplexity=30,n_iter=300)
tsne_results=tsne.fit_transform(X_scl)
print('t-SNE is completed, elapsed time is {}'.format(time.time()-time_start))
```

```python
plt.figure(figsize=(10,5))
plt.scatter(tsne_results[:, 0], tsne_results[:, 1],c=y)
plt.xticks([])
plt.yticks([])
plt.axis('off')
plt.show()
```

```python
time_start=time.time()
tsne=TSNE(n_components=2,verbose=1,perplexity=50,n_iter=300)
tsne_results=tsne.fit_transform(X_scl)
print('t-SNE is completed, elapsed time is {}'.format(time.time()-time_start))
```

```python
plt.figure(figsize=(10,5))
plt.scatter(tsne_results[:, 0], tsne_results[:, 1],c=y)
plt.xticks([])
plt.yticks([])
plt.axis('off')
plt.show()
```

We can guess and distinguish some clusters clearly, but majority of them are complicated.

```python
pca = PCA(n_components=2)
pca_components = pca.fit_transform(X_scl)
```

```python
plt.figure(figsize=(10,5))
plt.scatter(pca_components.T[0],pca_components.T[1],c=y)
plt.show()
```

As we seen above,t-SNE is better than PCA. Although this situation, t-SNE is not enough to distinguish all of the clusters.
