```python
print('AVG ile balayan stunlar ok fazla boluk bulundurduu iin kaldrabilirim.\nDier stunlarda yine ortalama ve ya yl deerleri olduu iin enterpolasyon yaparak doldurabilirim.')
```

```python
columns =      ["ENROLL", "TOTAL_REVENUE", "FEDERAL_REVENUE",
             "STATE_REVENUE", "LOCAL_REVENUE", "TOTAL_EXPENDITURE",
             "INSTRUCTION_EXPENDITURE", "SUPPORT_SERVICES_EXPENDITURE",
             "OTHER_EXPENDITURE", "CAPITAL_OUTLAY_EXPENDITURE", "GRADES_PK_G",
             "GRADES_KG_G", "GRADES_4_G", "GRADES_8_G", "GRADES_12_G", "GRADES_1_8_G",
             "GRADES_9_12_G", "GRADES_ALL_G"]
```

```python
for i in columns:
    df2[i].fillna(df2[i].mean(),inplace=True)
```

```python
pd.concat([df2.isnull().sum(), 100 * df2.isnull().sum()/len(df2)], 
              axis=1).rename(columns={0:'Missing Records', 1:'Percentage (%)'})
```

```python
df2['STATE'].unique()
```

```python
df3=df.copy()
cols=["ENROLL","TOTAL_REVENUE", "FEDERAL_REVENUE","STATE_REVENUE"]
```

```python
df3['YEAR'].unique()
```

```python
for i in cols:
    for j in df3['YEAR'].unique():
        df3.loc[df3['YEAR']==j,i].fillna(df3[df3["YEAR"]==j][i].mean(),inplace=True)
df3
```

```python
pd.concat([df3.isnull().sum(), 100 * df3.isnull().sum()/len(df3)], 
              axis=1).rename(columns={0:'Missing Records', 1:'Percentage (%)'})
```

```python
df4=df.copy()

for col in columns:
    df4.loc[:, col].interpolate(limit_direction='both',inplace=True)
```

```python
df4
```

```python
pd.options.display.float_format = '{:.2f}'.format
for col in columns:
    display(pd.concat([df[col],df2[col],df3[col],df4[col]],axis=1).describe())
```
