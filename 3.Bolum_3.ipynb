```python
google=pd.read_csv('../GOOG.csv')
google
```

```python
hissesenedi=[]
for x in range(len(google)-1):
    h=(google.Open[x+1]-google.Open[x])/google.Open[x]
    hissesenedi.append(h)
print(hissesenedi)
```

```python
mu=np.mean(hissesenedi)
sigma=np.std(hissesenedi,ddof=1)
y=sns.distplot(hissesenedi)
plt.show()
```
